```python
# Remove non-numeric characters from "Age" column
df_train['Age'] = df_train['Age'].str.replace(r'[^0-9-]', '', regex=True)
df_test['Age'] = df_test['Age'].str.replace(r'[^0-9-]', '', regex=True)

# Convert "Age" column to numeric
df_train['Age'] = pd.to_numeric(df_train['Age'], errors='coerce').astype('Int64')
df_test['Age'] = pd.to_numeric(df_test['Age'], errors='coerce').astype('Int64')

# Replace negative values in "Age" column with 0
df_train['Age'] = df_train['Age'].apply(lambda x: max(0, x))
df_test['Age'] = df_test['Age'].apply(lambda x: max(0, x))

# Replace ages greater than 80 with NaN
df_train['Age'] = df_train['Age'].apply(lambda x: x if 0 <= x <= 80 else np.nan)
df_test['Age'] = df_test['Age'].apply(lambda x: x if 0 <= x <= 80 else np.nan)

# Replace NaN values with 0 and convert to Int64
df_train['Age'] = df_train['Age'].fillna(0).astype('Int64')
df_test['Age'] = df_test['Age'].fillna(0).astype('Int64')
```
